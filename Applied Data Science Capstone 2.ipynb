```python
grouped_df = pd.read_csv('data/toronto-postalcode.csv')
grouped_df.drop(["Unnamed: 0"], axis=1, inplace=True)

print('The postalcode dataframe shape is', grouped_df.shape)
grouped_df.head()
```

```python
coordinates_df = pd.read_csv('data/Geospatial_Coordinates.csv')
print('The coordinates dataframe shape is', coordinates_df.shape)
coordinates_df.head()
```

```python
# join data frame
postcodes_with_coordinates_df = grouped_df.join(coordinates_df.set_index('Postal Code'), on='Postalcode')

print('The postcodes with coordinates dataframe shape is', postcodes_with_coordinates_df.shape)
postcodes_with_coordinates_df.to_csv('data/toronto-postalcode-latlong.csv')
postcodes_with_coordinates_df.head(12)
```
