```python
df_raw = pd.read_csv('data/id_db_latlong_data.csv', delimiter=';')

print('The latlong dataframe shape is', df_raw.shape)
df_raw.head()
```

do data preparation for coordinate data

```python
df_raw.drop(
    df_raw[['admin1','admin2','admin3','admin4','ac_code','elevation','population','dem','m_date','f_class']], 
    axis=1, 
    inplace=True
)
df_raw.drop(
    df_raw[['f_code','c_code']], 
    axis=1, 
    inplace=True
)

print('The latlong dataframe shape is', df_raw.shape)
df_raw.head()
```

```python
df_raw = df_raw.loc[df_raw['asciiname'].str.contains('Kelurahan') & (df_raw['timezone']=='Asia/Jakarta')]
df_raw['urban'] = df_raw.name.apply(lambda x: x[10:50])
df_raw['urban'] = df_raw['urban'].str.upper()
df_raw.to_csv('data/jakarta-latlong-final.csv')

print('The latlong dataframe shape is', df_raw.shape)
df_raw.head()
```

```python
df_latlong = pd.read_csv('data/jakarta-latlong-final.csv')
df_latlong.dropna()
df_latlong = df_latlong.reset_index()
df_latlong.drop(df_latlong[['Unnamed: 0','geonameid','index','alternatenames']],axis=1, inplace=True)

print('The latlong dataframe shape is', df_latlong.shape)
df_latlong.head()
```
